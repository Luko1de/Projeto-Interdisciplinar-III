```python
features = [
    'popularity',
    'budget',
    'revenue',
    'runtime',
    'vote_average',
    'vote_count',
    'ROI',
    'released_year',
    'recommendations_count',
    'status_b',
    'sucesso',
    'fracasso',
    'ROI',
    'Action',
    'Adventure',
    'Animation',
    'Comedy',
    'Crime',
    'Documentary',
    'Drama',
    'Family',
    'Fantasy',
    'History',
    'Horror',
    'Music',
    'Mystery',
    'Romance',
    'Science Fiction',
    'TV Movie',
    'Thriller',
    'War',
    'Western'
]

targets = ['sucesso', 'fracasso']

X = df[features]
y = df[targets]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
def count_outcomes(df):
    print("Quantidade de resultados:")
    
    count_success = df['sucesso'].value_counts()
    print("Sucesso:")
    print(f"Sucesso: {count_success.get(1, 0)}")  # Contagem de sucesso
    print(f"Não sucesso: {count_success.get(0, 0)}\n")  # Contagem de fracasso

    count_failure = df['fracasso'].value_counts()
    print("Fracasso:")
    print(f"Fracasso: {count_failure.get(1, 0)}")  # Contagem de fracasso
    print(f"Não fracasso: {count_failure.get(0, 0)}\n")  # Contagem de sucesso

# Chamar a função
count_outcomes(df)
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import pandas as pd

# Verificar a distribuição das classes no conjunto de treino e teste
print("Distribuição das classes no conjunto de treino:")
print(y_train.value_counts())

print("\nDistribuição das classes no conjunto de teste:")
print(y_test.value_counts())

# Garantir que y_train seja uma Série (vetor unidimensional)
if isinstance(y_train, pd.DataFrame):
    y_train = y_train.squeeze()  # Converte para uma Série se for um DataFrame

# Verificar tipos de dados
print("\nTipos de dados em X_train:")
print(X_train.dtypes)

# Aplicar SMOTE apenas no conjunto de treino
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Verificar a distribuição das classes após o SMOTE
print("\nDistribuição das classes no conjunto de treino após o SMOTE:")
print(y_train_balanced.value_counts())
```

Classificação por três algoritmos

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import numpy as np

# Definir os rótulos personalizados conforme suas classes
target_names = ['sucesso', 'fracasso']  # Substitua pelos nomes adequados

# Treinar o modelo Random Forest com o conjunto de treino balanceado (após o SMOTE)
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = rf_classifier.predict(X_test)

# Exibir o relatório de classificação com os rótulos personalizados
print("Relatório de Classificação (Random Forest):")
print(classification_report(y_test, y_pred, target_names=target_names))

# Exibir a matriz de confusão formatada com os rótulos personalizados
print("Matriz de Confusão (Random Forest):")
conf_matrix = confusion_matrix(y_test, y_pred)

# Mostrar a matriz de confusão formatada com os rótulos personalizados para melhor visualização
conf_matrix_df = pd.DataFrame(conf_matrix, index=target_names, columns=target_names)
print(conf_matrix_df)
```
